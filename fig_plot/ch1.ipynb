```python
acc_vgg11 = [0.72, 0.763529, 0.844, 0.849231, 0.892727, 0.9, 0.891429, 0.92]
acc_resnet18 = [0.704211, 0.802353, 0.834667, 0.823077, 0.861818, 0.9, 0.888571, 0.904]
acc_resnet50 = [0.717895, 0.845882, 0.878667, 0.863077, 0.870909, 0.937778, 0.937142, 0.916]
acc_efficientnet = [0.651579, 0.824706, 0.857333, 0.856923, 0.852727 ,0.868889, 0.868571, 0.908]
acc_mobilenet = [0.653684, 0.805882, 0.852, 0.861538, 0.869091, 0.926667, 0.9, 0.916]

alpha_figsize = 7
fig, ax1=plt.subplots(1, 1, figsize=(1*alpha_figsize, 0.618*alpha_figsize), dpi=200)
ax1.plot([x*100 for x in acc_vgg11[::-1]], marker='D', label="VGG11")
ax1.plot([x*100 for x in acc_resnet18[::-1]], marker='*', label="ResNet18")
ax1.plot([x*100 for x in acc_resnet50[::-1]], marker='+', label="ResNet50")
ax1.plot([x*100 for x in acc_efficientnet[::-1]], marker='x', label="EfficientNet")
ax1.plot([x*100 for x in acc_mobilenet[::-1]], marker='o', label="MobileNet")

ax1.set_xlabel("每类训练样本数量", fontsize=13)
ax1.set_xticks(np.arange(0, 8, 1))
ax1.set_xticklabels(['10-shot', '30-shot', '50-shot', '70-shot', '90-shot', '110-shot', '130-shot', '150-shot'][::-1], fontdict={'fontweight':'bold'})

ax1.set_ylabel("准确率Acc(%)", fontsize=13)
ax1.set_ylim(60, 100)

plt.grid(axis='y', alpha=0.7)

plt.legend()
plt.show()
```
