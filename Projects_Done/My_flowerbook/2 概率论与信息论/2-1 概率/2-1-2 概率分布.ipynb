* ***概率质量函数***（*probability mass function*）：表示离散型变量的概率分布  
* ***概率密度函数***（*probability density function*）：表示连续性变量的概率分布  
* ***累计分布函数***（*cummulative distribution function*）：表示小于$x$的概率的积分

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform
%matplotlib inline

# 生成样本
fig, ax = plt.subplots(1, 1)
r = uniform.rvs(loc=0, scale=1, size=1000) # 生成1000个[0, 1]上满足均匀分布的数
ax.hist(r, density=True, histtype='stepfilled', alpha=0.5)

# 均匀分布的PDF
x = np.linspace(uniform.ppf(0.01), uniform.ppf(0.99), 100)
ax.plot(x, uniform.pdf(x), 'r-', lw=5, alpha=0.8, label='uniform pdf')
```
