```python
%matplotlib inline
import torch
from d2l import torch as d2l
```

```python
# ReLUå‡½æ•°
x=torch.arange(-8.0, 8.0, 0.1, requires_grad=True)
y=torch.relu(x)
d2l.plot(x.detach(), y.detach())
```

```python
# ReLUå‡½æ•°çš„å¯¼æ•°
y.backward(torch.ones_like(x), retain_graph=True)
d2l.plot(x.detach(), x.grad,figsize=(5, 2.5))
```

```python
# sigmoidå‡½æ•°
y = torch.sigmoid(x)
d2l.plot(x.detach(), y.detach())
```

```python
# sigmoidå‡½æ•°æ¢¯åº¦
x.grad.data.zero_()
y.backward(torch.ones_like(x))
d2l.plot(x.detach(),x.grad)
```

```python
# tanhå‡½æ•°
y=torch.tanh(x)
d2l.plot(x.detach(),y.detach())
```

```python
# tanhæ¢¯åº¦
x.grad.data.zero_()
y.backward(torch.ones_like(x))
d2l.plot(x.detach(),x.grad)
```
